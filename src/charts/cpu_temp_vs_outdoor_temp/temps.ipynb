```python
import json
from datetime import datetime
import pandas as pd
```

```python
cpu = json.load(open("temperatures_cpu.json"))

cpux = [datetime.strptime(r["timestamp"], "%Y-%m-%d %H:%M:%S.%fZ") for r in cpu]
cpuy = [float(r["sensors"]) for r in cpu]
cpu_df = pd.DataFrame({"time": cpux, "temp": cpuy})
print(cpu_df)
```

```python
import numpy as np
arr = np.loadtxt("outdoor.csv",
				delimiter=";", dtype=str)

tmpx = [datetime.strptime(t, "%d.%m.%Y %H:%M") for t in arr[1:,1]]
tmpy = [float(t) for t in arr[1:,2]]
temp_df = pd.DataFrame({"time": tmpx, "temp": tmpy})
print(temp_df)
```

```python
import matplotlib.pyplot as plt

temp_df_filtered = temp_df[temp_df["time"] > datetime(day=28, month=5, year=2026)]

plt.figure(figsize=(15,10))
plt.subplot(211)
plt.plot(temp_df_filtered["time"], temp_df_filtered["temp"].rolling(window=144).mean(), label="10min instant with 24h rolling window")
plt.plot(temp_df_filtered["time"], temp_df_filtered["temp"].rolling(window=6).mean(), label="10min instant with 1h rolling window", alpha=0.5)
plt.xlim(datetime(day=1, month=6, year=2026), datetime.now())
plt.title("outdoor temperatures")
plt.ylabel("C°")
plt.legend()

plt.subplot(212)
plt.plot(cpu_df["time"], cpu_df["temp"].rolling(window=3).mean(), label="8h average with 24h rolling window")
plt.xlim(datetime(day=1, month=6, year=2026), datetime.now())
plt.title("indoor raspberry pi cpu temperatures")
plt.ylabel("C°")
plt.legend()
plt.show()
```
