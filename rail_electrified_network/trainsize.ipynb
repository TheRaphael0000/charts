```python
from matplotlib import pyplot as plt
from adjustText import adjust_text

# https://en.wikipedia.org/wiki/List_of_countries_by_rail_transport_network_size (17/11/2024)
rows = open("data.csv").read().split("\n")
rows = [r.split("\t") for r in rows]
rows = [r for r in rows if len(r) == 4]
ignored = 50
rows = [r for r in rows if float(r[2]) > ignored]

label = [r[0] for r in rows]
size = [float(r[1]) for r in rows]
electrified = [float(r[2]) for r in rows]
percent = [float(r[3]) for r in rows]

plt.figure(figsize=(10,8), dpi=500)
plt.grid()
plt.scatter(percent, size)
plt.title("Rail transport network size over percent electrified")
plt.xlabel("Percent of electrified network length (%)")
plt.ylabel("Total network length (km)")
plt.yscale("log")

texts = []
for i in range(len(percent)):
    t = plt.text(percent[i], size[i], label[i])
    texts.append(t)

adjust_text(texts, arrowprops=dict(arrowstyle="-", color='k', lw=0.5))
plt.show()
```
